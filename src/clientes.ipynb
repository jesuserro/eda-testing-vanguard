# Análisis de las columnas del archivo clients.csv

En este notebook, analizaremos en profundidad las columnas del archivo `clients.csv`, describiendo su tipo y los valores que contienen. Esto nos ayudará a entender mejor los datos y cómo podemos utilizarlos en nuestro análisis.

## Descripción de las 9 columnas

1. **client_id**: Identificador único del cliente. Es un valor numérico entero.
2. **clnt_tenure_yr**: Antigüedad del cliente en años. Es un valor numérico entero.
3. **clnt_tenure_mnth**: Antigüedad del cliente en meses. Es un valor numérico entero.
4. **clnt_age**: Edad del cliente en años. Es un valor numérico decimal.
5. **gendr**: Género del cliente. Es un valor categórico que puede ser 'M' (Masculino), 'F' (Femenino) o 'U' (Desconocido).
6. **num_accts**: Número de cuentas que tiene el cliente. Es un valor numérico entero.
7. **bal**: Saldo total en las cuentas del cliente. Es un valor numérico decimal.
8. **calls_6_mnth**: Número de llamadas realizadas por el cliente en los últimos 6 meses. Es un valor numérico entero.
9. **logons_6_mnth**: Número de veces que el cliente ha iniciado sesión en los últimos 6 meses. Es un valor numérico entero.

```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
file_path = '../data/processed/clients.csv'
clients_df = pd.read_csv(file_path)

# Mostrar las primeras filas del DataFrame
clients_df.head()
```

## Correlación de variables numéricas

```python
# Capturar las columnas numéricas
df_numerical = clients_df.select_dtypes(include=['int64', 'float64'])

# IMPORTANT: The Spearman correlation coefficient ranges from -1 to 1, where:
correlation_matrix = df_numerical.corr()

# Setting up the matplotlib figure with an appropriate size
plt.figure(figsize=(18, 15))

# Drawing the heatmap for the numerical columns
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap for Selected Numerical Variables")
plt.show()
```
